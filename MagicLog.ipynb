```python
import matplotlib.pyplot as plt
import numpy as np

# Data points from the image
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
exportaciones = np.array([21833.3, 18393.2, 14174.0, 9896.2, 10250.0, 10615.2, 10693.2, 11320.1, 8910.4, 10880.4, 14758.5, 13081.6])
importaciones = np.array([14213.7, 16336.7, 18192.6, 15512.4, 11877.9, 12014.5, 12986.0, 13276.8, 10535.0, 14046.2, 18839.1, 15997.7])

# Plotting the grouped bar chart
width = 0.35  # width of the bars
fig, ax = plt.subplots()
ax.bar(years - width/2, exportaciones, width, label='Exportaciones', color='skyblue')
ax.bar(years + width/2, importaciones, width, label='Importaciones', color='orange')

ax.set_xlabel('Año')
ax.set_ylabel('Millones de USD')
ax.set_title('Exportaciones vs Importaciones (2012-2023)')
ax.set_xticks(years)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points from the image
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
exportaciones = np.array([21833.3, 18393.2, 14174.0, 9896.2, 10250.0, 10615.2, 10693.2, 11320.1, 8910.4, 10880.4, 14758.5, 13081.6])
importaciones = np.array([14213.7, 16336.7, 18192.6, 15512.4, 11877.9, 12014.5, 12986.0, 13276.8, 10535.0, 14046.2, 18839.1, 15997.7])

# Colors from the logo (approximated)
export_color = '#00AEEF'  # Light blue
import_color = '#003366'  # Darker blue

# Plotting the grouped bar chart
width = 0.35  # width of the bars
fig, ax = plt.subplots()

fig.set_size_inches(20, 10)

# Create the bars
bars1 = ax.bar(years - width/2, exportaciones, width, label='Exportaciones', color=export_color)
bars2 = ax.bar(years + width/2, importaciones, width, label='Importaciones', color=import_color)

# Add the totals on top of each bar
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 500, f'{yval:.1f}', ha='center', va='bottom', fontsize=9)

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 500, f'{yval:.1f}', ha='center', va='bottom', fontsize=9)

# Set labels and title
ax.set_xlabel('Año')
ax.set_ylabel('Millones de USD')
ax.set_title('Exportaciones vs Importaciones (2012-2023)')
ax.set_xticks(years)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Years
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])

# Data points with the $ symbol and no decimals
exportaciones = np.array(['$21,833', '$18,393', '$14,174', '$9,896', '$10,250', '$10,615', '$10,693', '$11,320', '$8,910', '$10,880', '$14,758', '$13,081'])
importaciones = np.array(['$14,213', '$16,336', '$18,192', '$15,512', '$11,877', '$12,014', '$12,986', '$13,276', '$10,535', '$14,046', '$18,839', '$15,997'])

# Numeric values for plotting
export_values = np.array([21833, 18393, 14174, 9896, 10250, 10615, 10693, 11320, 8910, 10880, 14758, 13081])
import_values = np.array([14213, 16336, 18192, 15512, 11877, 12014, 12986, 13276, 10535, 14046, 18839, 15997])

# Plotting the grouped bar chart
width = 0.4  # width of the bars, made wider
space_between_bars = 0.05  # space between the bars for the same year
fig, ax = plt.subplots()

fig.set_size_inches(25, 10)

# Create bars with space between them
bars1 = ax.bar(years - width/2 - space_between_bars/2, export_values, width, label='Exportaciones', color='#00AEEF')
bars2 = ax.bar(years + width/2 + space_between_bars/2, import_values, width, label='Importaciones', color='#003366')

# Add the totals on top of each bar with diagonal rotation and $ symbol
for bar, label in zip(bars1, exportaciones):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 300, label, ha='center', va='bottom', fontsize=14, rotation=45)

for bar, label in zip(bars2, importaciones):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 300, label, ha='center', va='bottom', fontsize=14, rotation=45)

# Set labels and title
ax.set_xlabel('Año', fontsize=16)
ax.set_ylabel('Millones de USD', fontsize=16)
ax.set_title('Exportaciones vs Importaciones (2012-2023)', fontsize=20)
ax.set_xticks(years)
ax.legend(fontsize=14)

# Remove borders and grid lines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.grid(False)  # Remove grid lines

plt.show()
```
