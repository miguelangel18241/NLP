# **Assessment 1 NLP**
## Exercise 1
### Named Entity Recognition: NER


````python
from transformers import pipeline

# Define your text for NER
text_for_ner = """
Nvidia Corporation, headquartered in Santa Clara, California, has been a leader in the semiconductor industry for decades. Founded by Jensen Huang, Nvidia has revolutionized the gaming and artificial intelligence sectors with its cutting-edge graphics processing units (GPUs). The company's recent financial performance has been stellar, with a significant surge in stock price following robust earnings reports. Nvidia's major competitors include Advanced Micro Devices (AMD) and Intel Corporation, both of which are also based in Silicon Valley.

In addition to its core business in gaming and AI, Nvidia has expanded into autonomous vehicles, data centers, and healthcare technologies. The company collaborates closely with research institutions and universities worldwide to advance AI research and development. Its CEO, Jensen Huang, is known for his visionary leadership and strategic acquisitions that have propelled Nvidia to the forefront of technological innovation.

Nvidia's global footprint extends beyond the United States, with major offices in countries such as China, India, and Germany. The company continues to invest heavily in research and development to maintain its competitive edge in an increasingly digital world.
"""

# Initialize NER pipeline
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", revision="f2482bf")

# Perform NER
ner_results = ner_pipeline(text_for_ner)

# Print the identified entities
entities = [(x['word'], x['entity']) for x in ner_results]
if entities:
    print("Entities found:")
    for x, entity_type in entities:
        print(f"{x}: {entity_type}")
else:
    print("No entities found in the specified classes.")
````

# **Assessment 1 NLP**
## Exercise 2
### Sentiment Analysis

````python
# Define your text for Sentiment Analysis
text = """
Nvidia Gets Rare Downgrade on Valuation Concerns After Rally. Nvidia Corporation, headquartered in Santa Clara, California, has been a leader in the semiconductor industry for decades. Founded by Jensen Huang, Nvidia has revolutionized the gaming and artificial intelligence sectors with its cutting-edge graphics processing units (GPUs). The company's recent financial performance has been stellar, with a significant surge in stock price following robust earnings reports. Nvidia's major competitors include Advanced Micro Devices (AMD) and Intel Corporation, both of which are also based in Silicon Valley.

In addition to its core business in gaming and AI, Nvidia has expanded into autonomous vehicles, data centers, and healthcare technologies. The company collaborates closely with research institutions and universities worldwide to advance AI research and development. Its CEO, Jensen Huang, is known for his visionary leadership and strategic acquisitions that have propelled Nvidia to the forefront of technological innovation.

Nvidia's global footprint extends beyond the United States, with major offices in countries such as China, India, and Germany. The company continues to invest heavily in research and development to maintain its competitive edge in an increasingly digital world.
"""
````

````python
# I need to import nltk to be able to make sentences after each full stop.
!pip install nltk
````

````python
import nltk
nltk.download('punkt')
````

````python
# Tokenize the text into sentences

sentences = nltk.sent_tokenize(text)
for sentence in sentences:

  print(sentence)
````

````python
index_sentence = 0
for word in sentences:
  index_sentence = index_sentence + 1
  words = word.split()
  print(f"Sentence {index_sentence} has {len(words)} words.")
  print(words)
  print()
````

##Sentences for analysis
###I have split my text to be able to use every sentence for the Sentiment Analysis.

````python
total_number_of_words = 0
for index, sentence in enumerate(sentences, start=1):
  words = sentence.split()
  num_words = len(words)
  total_number_of_words += num_words
  print(f"Sentence {index} has {words}.")
  print(words)
  print()
print(f"Total sentences: {len(sentences)}, and total number of words is: {total_number_of_words}")
````

````python

#Pipeline for sentiment analysis.
## Initialize it

sentiment_pipeline = pipeline("sentiment-analysis")
````

````python
for index, sentence in enumerate(sentences, start = 1):
    result = sentiment_pipeline(sentence)
    print(f"Sentence {index}: {sentence}")
    print(f"Sentiment: {result[0]['label']}, Score: {result[0]['score']}")
    print()
````
